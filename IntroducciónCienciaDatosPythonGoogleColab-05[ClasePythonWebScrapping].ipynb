# **03. Extraer Texto De HTML Con Métodos String**
[[Contenidos]](#contents)

Una forma de extraer información del HTML de una página web es utilizar métodos de la clase `string`. Por ejemplo, puede usar `.find()` para buscar en el texto del HTML las etiquetas `<title>` y extraer el título de la página web.

Extraigamos el título de la página web que solicitó en el ejemplo anterior. Si conoce el índice del primer carácter del título y el primer carácter de la etiqueta `</title>` de cierre, puede utilizar un segmento de cadena para extraer el título.

Dado que `.find()` devuelve el índice de la primera aparición de una subcadena, puede obtener el índice de la etiqueta `<title>` de apertura pasando la cadena `"<title>"` a `.find()`

````python
title_index = html.find("<title>")
title_index
````

Sin embargo, no desea el índice de la etiqueta <title>. Quieres el índice del título en sí. Para obtener el índice de la primera letra del título, puede agregar la longitud de la cadena `"<title>"` a `title_index`

````python
start_index = title_index + len("<title>")
start_index
````

Ahora obtenga el índice de la etiqueta `</title>` de cierre pasando la cadena `"</title>"` a `.find()`

````python
end_index = html.find("</title>")
end_index
````

Finalmente, puede extraer el título cortando la cadena html:

````python
title = html[start_index:end_index]
title
````

En algunos casos las páginas web pueden contener errores de codificacion y espacios que no permiten la correcta interpretacion de titulos, objetos y otros datos, para estos casos es neceario formas mas confiables para extraer texto de una página web.

# **04. Un Manual Sobre Expresiones Regulares**
[[Contenidos]](#contents)

Expresiones Regulares o Regular expressions mas conocidas como `regexes` son patrones que se pueden usar para buscar texto dentro de una cadena. Python admite expresiones regulares a través del módulo [re](https://docs.python.org/3/library/re.html) de la biblioteca estándar.


*Nota: Las expresiones regulares no son específicas de Python. Son un concepto de programación general y se pueden usar con cualquier lenguaje de programación.*

Para trabajar con expresiones regulares, lo primero que debe hacer es importar el módulo re:



```python
import re
```

Las expresiones regulares usan caracteres especiales llamados metacaracteres para denotar diferentes patrones. Por ejemplo, el carácter de asterisco `(*)` representa cero o más de lo que venga justo antes del asterisco.

En el siguiente ejemplo, usa `findall()` para buscar cualquier texto dentro de una cadena que coincida con una expresión regular determinada:





````python
import re
re.findall("ab*c", "ac")
````

El primer argumento de `re.findall()` es la expresión regular que desea hacer coincidir, y el segundo argumento es la cadena a probar. En el ejemplo anterior, busca el patrón `"ab*c"` en la cadena `"ac"`.

La expresión regular `"ab*c"` coincide con cualquier parte de la cadena que comienza con una "a", termina con una "c" y tiene cero o más instancias de `"b"` entre las dos. `re.findall()` devuelve una lista de todas las coincidencias. La cadena `"ac"` coincide con este patrón, por lo que se devuelve en la lista.

Este es el mismo patrón aplicado a diferentes cadenas:


````python
re.findall("ab*c", "abcd")
````

````python
re.findall("ab*c", "acc")
````

````python
re.findall("ab*c", "abcac")
````

````python
re.findall("ab*c", "abdc")
````

Tenga en cuenta que si no se encuentra ninguna coincidencia, `findall()` devuelve una lista vacía.

La coincidencia de patrones distingue entre mayúsculas y minúsculas. Si desea hacer coincidir este patrón independientemente del caso, puede pasar un tercer argumento con el valor `re.IGNORECASE`

````python
re.findall("ab*c", "ABC")
````

````python
re.findall("ab*c", "ABC", re.IGNORECASE)
````

Puede utilizar un punto `(.)` para representar cualquier carácter en una expresión regular. Por ejemplo, puede encontrar todas las cadenas que contienen las letras `"a"` y `"c"` separadas por un solo carácter de la siguiente manera.


````python
re.findall("a.c", "abc")
````

````python
re.findall("a.c", "abbc")
````

````python
re.findall("a.c", "ac")
````

````python
re.findall("a.c", "acc")
````

El patrón `.*` en una expresión regular representa cualquier carácter repetido cualquier número de veces. Por ejemplo, `"a.*C"` se puede usar para encontrar cada subcadena que comience con "a" y termine con "c", independientemente de la letra o letras que se encuentren entre ellas

````python
re.findall("a.*c", "abc")
````

````python
re.findall("a.*c", "abbc")
````

````python
re.findall("a.*c", "ac")
````

````python
re.findall("a.*c", "acc")
````

`re.search()` se usa para buscar un patrón particular dentro de una cadena. Esta función es algo más complicada que `re.findall()` porque devuelve un objeto llamado `MatchObject` que almacena diferentes grupos de datos. Esto se debe a que puede haber coincidencias dentro de otras coincidencias y `re.search()` devuelve todos los resultados posibles.

Los detalles del `Match Object` son irrelevantes en el presente contexto, puede profundizar este conocimiento en la siguiente ruta [`Match Object`](https://docs.python.org/3/library/re.html#re.search). Por ahora, nos interesa el contexto para usar `.group()` en un MatchObject el cual devolverá el primer resultado y el más inclusivo, lo cual, en la mayoría de los casos es justo lo que desea:

````python
match_results = re.search("ab*c", "ABC", re.IGNORECASE)
match_results.group()
````

Hay una función más en el módulo `re` que es útil para analizar el texto. `re.sub()`, que es la abreviatura de sustituto (substitute), le permite reemplazar texto en una cadena que coincide con una expresión regular con texto nuevo. Se comporta como el método de cadena `.replace()`.

Los argumentos pasados a `re.sub()` son la expresión regular, seguida del texto de reemplazo, seguido de la cadena.


re.sub () usa la expresión regular `"<. *>"` para buscar y reemplazar todo entre el primer `<y el último>`, que se extiende desde el principio de `<replaced>` hasta el final de `<tags>`. Esto se debe a que las expresiones regulares de Python son codiciosas (**greedy**), lo que significa que intentan encontrar la coincidencia más larga posible cuando se utilizan caracteres como `*`.

Alternativamente, puede usar el patrón de coincidencia no codicioso `*?`, que funciona de la misma manera que `*` excepto que coincide con la cadena de texto más corta posible, esta vez, `re.sub()` encuentra dos coincidencias, `<replaced>` y `<tags>`, y sustituye la cadena "ELEPHANTS" para ambas coincidencias. 

````python
string = "Everything is <replaced> if it's in <tags>."
string = re.sub("<.*>", "ELEPHANTS", string)
string
````

````python
string = "Everything is <replaced> if it's in <tags>."
string = re.sub("<.*?>", "ELEPHANTS", string)
string
````

# **05. Extraer Texto De HTML Con Expresiones Regulares**
[[Contenidos]](#contents)

Intentemos ahora analizar el título de una nueva página, que incluye líneas de HTML de manera erratica y con problemas de contexto.
*   http://olympus.realpython.org/profiles/dionysus



```html
<TITLE >Profile: Dionysus</title  / >
```

El método `.find()` tendría dificultades para lidiar con las inconsistencias aquí, pero con el uso inteligente de expresiones regulares, puede manejar este código de manera rápida y eficiente:



````python
import re
from urllib.request import urlopen

url = "http://olympus.realpython.org/profiles/dionysus"
page = urlopen(url)
html = page.read().decode("utf-8")

pattern = "<title.*?>.*?</title.*?>"
match_results = re.search(pattern, html, re.IGNORECASE)
title = match_results.group()
title = re.sub("<.*?>", "", title) # Remove HTML tags

print(title)
````

Detallemos el algoritmo en donde la primera expresión regular en la cadena del patrón dividiéndola en tres partes:

1.   `<title.*?>` coincide con la etiqueta de apertura `<TITLE>` en html. La parte `<title` del patrón coincide con `<TITLE` porque `re.search()` se llama con `re.IGNORECASE`, y. `*?>` Coincide con cualquier texto después de `<TITLE` hasta la primera instancia de `>`.

2.   `.*?` coincide sin avidez con todo el texto después de la apertura `<TITLE>`, deteniéndose en la primera coincidencia de `</title.*?>.`

3.   `</title.*?> ` difiere del primer patrón solo en el uso del carácter `/`, por lo que coincide con la etiqueta `</title / >` de cierre en html.

La segunda expresión regular, la cadena `"<.*?>"`, También usa la expresión no codiciosa `.*?` para que coincida con todas las etiquetas HTML en la cadena de título. Al reemplazar cualquier coincidencia con "", `re.sub()` elimina todas las etiquetas y devuelve solo el texto.

Nota: Web scraping en Python o en cualquier otro idioma puede resultar tedioso. No hay dos sitios web que estén organizados de la misma manera y HTML suele ser complicado. Además, los sitios web cambian con el tiempo. No se garantiza que los web scrapers que funcionan hoy funcionen el año que viene, ¡o la semana que viene!

# **08. Crear Un Objeto Beautifulsoup**
[[Contenidos]](#contents)

Escriba el siguiente programa para crear un objeto tipo Beautifulsoup.

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "http://olympus.realpython.org/profiles/dionysus"
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```
Este programa hace tres cosas:

1.   Abre la URL `http://olympus.realpython.org/profiles/dionysus` usando `urlopen()` del módulo `urllib.request`.

2.   Lee el HTML de la página como una cadena y lo asigna a la variable html

3.   Crea un objeto `BeautifulSoup` y lo asigna a la variable `soup`.

El objeto BeautifulSoup asignado a soup se crea con dos argumentos. El primer argumento es el HTML que se va a analizar, y el segundo argumento, la cadena `"html.parser"`, le dice al objeto qué analizador utilizar detrás de escena. `"html.parser"` representa el analizador HTML integrado de Python.




````python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "http://olympus.realpython.org/profiles/dionysus"
#url = "https://webscraper.io/test-sites/e-commerce/allinone"
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
````

# **09. Usar Un Objeto Beautifulsoup**
[[Contenidos]](#contents)

Los objetos BeautifulSoup tienen un método `.get_text()` que se puede usar para extraer todo el texto del documento y eliminar automáticamente cualquier etiqueta HTML.

Detallemos en un ejemplo el método `.get_text()`.

```python
print(soup.get_text())
```
Hay muchas líneas en blanco en esta salida. Estos son el resultado de caracteres de nueva línea en el texto del documento HTML. Puede eliminarlos con el método string `.replace()` si es necesario.

A menudo, necesita obtener solo texto específico de un documento HTML. Usar Beautiful Soup primero para extraer el texto y luego usar el método de cadena `.find()` es a veces más fácil que trabajar con expresiones regulares.

Sin embargo, a veces las propias etiquetas HTML son los elementos que señalan los datos que desea recuperar. Por ejemplo, quizás desee recuperar las URL de todas las imágenes de la página. Estos enlaces están contenidos en el atributo `src` de las etiquetas HTML `<img>`.

En este caso, puede usar `find_all()` para devolver una lista de todas las instancias de esa etiqueta en particular.

```python
soup.find_all("img")
```
Esto devuelve una lista de todas las etiquetas `<img>` en el documento HTML. Los objetos de la lista parecen ser cadenas que representan las etiquetas, pero en realidad son instancias del objeto Etiqueta proporcionado por Beautiful Soup. Los objetos de etiqueta proporcionan una interfaz sencilla para trabajar con la información que contienen.

Exploremos esto un poco primero desempaquetando los objetos Etiqueta de la lista como se muestra a continuación con el uso de ciclos `for`.

```python
print(soup.get_text())
print(len(soup.find_all("img")))
imagenes = []
for i in soup.find_all("img"):
    print(i.name)
    print(i["src"])
    imagenes.append(i)
for j in imagenes:
    print(j)
```
Puede acceder a los atributos HTML del objeto Tag poniendo su nombre entre corchetes, como si los atributos fueran claves en un diccionario.

Por ejemplo, la etiqueta `<img src ="/static/dionysus.jpg"/>` tiene un solo atributo, `src`, con el valor `"/static/dionysus.jpg"`. Asimismo, una etiqueta HTML como el enlace `<a href="https://realpython.com" target="_blank">` tiene dos atributos, href y target.

Para obtener la fuente de las imágenes en la página del perfil de Dionysus, acceda al atributo `src` usando la notación del diccionario mencionada anteriormente.

Se puede acceder a determinadas etiquetas en documentos HTML mediante las propiedades del objeto Etiqueta. Por ejemplo, para obtener la etiqueta `<title>` en un documento, puede usar la propiedad `.title`.

```python
soup.title
```

Beautiful Soup limpia automáticamente las etiquetas eliminando espacios adicionales y posibles errores en la etiqueta de apertura y la barra oblicua (/) en la etiqueta de cierre.

También puede recuperar solo la cadena entre las etiquetas de título con la propiedad .string del objeto Etiqueta:

```python
soup.title.string
```




````python
print(soup.get_text())
print(len(soup.find_all("img")))
print('+++++++++++++++++++-------------------+++++++++++++++++++')
imagenes = []
for i in soup.find_all("img"):
    print(i.name)
    print(i["src"])
    imagenes.append(i)
print('+++++++++++++++++++-------------------+++++++++++++++++++')
for j in imagenes:
    print(j)
print(soup.title)
print(soup.title.string)
````

Una de las características más útiles de Beautiful Soup es la capacidad de buscar tipos específicos de etiquetas cuyos atributos coincidan con ciertos valores. Por ejemplo, si desea buscar todas las etiquetas `<img>` que tienen un atributo `src` igual al valor `/static/dionysus.jpg`, puede proporcionar el siguiente argumento adicional a `.find_all()`.



```python
soup.find_all("img", src="/static/dionysus.jpg")
```
Este ejemplo es algo arbitrario y la utilidad de esta técnica puede no ser evidente en el ejemplo. Si pasa algún tiempo navegando por varios sitios web y viendo las fuentes de sus páginas, notará que muchos sitios web tienen estructuras HTML extremadamente complicadas. Al extraer datos de sitios web con Python, a menudo te interesan partes específicas de la página. Al dedicar un tiempo a examinar el documento HTML, puede identificar etiquetas con atributos únicos que puede utilizar para extraer los datos que necesita.

Luego, en lugar de depender de expresiones regulares complicadas o usar `.find()` para buscar en el documento, puede acceder directamente a la etiqueta en particular que le interesa y extraer los datos que necesita. En algunos casos, es posible que Beautiful Soup no ofrezca la funcionalidad que necesita. La biblioteca `lxml` es algo más complicada para empezar, pero ofrece mucha más flexibilidad que Beautiful Soup para analizar documentos HTML. Es posible que desee comprobarlo una vez que se sienta cómodo con Beautiful Soup.

**Nota**: Los analizadores HTML como Beautiful Soup pueden ahorrarle mucho tiempo y esfuerzo cuando se trata de localizar datos específicos en páginas web. Sin embargo, a veces HTML está tan mal escrito y desorganizado que incluso un analizador sofisticado como Beautiful Soup no puede interpretar las etiquetas HTML correctamente. En este caso, a menudo te quedas con el uso de `.find()` y técnicas de expresión regular para tratar de analizar la información que necesitas.

BeautifulSoup es ideal para extraer datos del HTML de un sitio web, pero no proporciona ninguna forma de trabajar con formularios HTML. Por ejemplo, si necesita buscar en un sitio web alguna consulta y luego obtener los resultados, BeautifulSoup por sí solo no lo llevará muy lejos.

# **10. Interactuar Con Formularios HTML**
[[Contenidos]](#contents)

## **Validador de URL's**

Primero, hagamos un validador de URL, que se asegure de que la URL pasada sea válida, ya que hay algunos sitios web que colocan datos codificados en lugar de una URL, por lo que debemos omitirlos.
La función `urlparse()` analiza una URL en seis componentes, solo necesitamos ver si el `netloc` (nombre de dominio) y el `scheme` (protocolo) están ahí.



````python
def url_valida(url):
    """
    Verifica que una URL sea valida.
    """
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)
````

### **El contenido HTML**

Nota: Todo el código descrito a continuación se realiza en la función previamente descrita.

El contenido HTML de la página web está en el objeto `soup`, para extraer todas las etiquetas img en HTML, necesitamos usar el método `soup.find_all("img")`.

Esto recuperará todos los elementos img en una lista de Python.

Sin embargo, está embedido en un objeto `tqdm` solo para imprimir una barra de progreso. Para obtener la URL de una etiqueta img, hay un atributo src. Sin embargo, hay algunas etiquetas que no contienen el atributo src, las omitimos usando la instrucción `continue`.

Hay algunas URL que contienen `HTTP GET key value` que no permiten obtener la imagen original, de tal manera que puenden terminar de la siguiente manera: 
*   `"/image.png?c=3.2.5"`
Procedemos a eliminar estos parametros o datos adicionales posteriores a la extension de la imagen.



```python
        try:
            pos = img_url.index("?")
            img_url = img_url[:pos]
        except ValueError:

```
Estamos obteniendo la posición del caracter `'?'`, luego quitando todo después de él, si no hay un caracter `'?'`, generará `ValueError`, es por eso que se utiliza el bloque `try / except`. A continuacion se deberá validar que todas las URL sean validas.




A continuacion la funcion completa.




````python
def Todas_Las_Imagenes(url):
    """
    Returna todas las imagenes URL en un solo URL
    """
    soup = bs(requests.get(url).content, "html.parser")
    urls = []
    for img in tqdm(soup.find_all("img"), "Extracting images"):
        img_url = img.attrs.get("src")
        if not img_url:
            # si la imagen no tiene un atributo src, salta la ejecución
            continue
        # se realiza la url completa, con dominio y subrutas de las imagenes.
        img_url = urljoin(url, img_url)
        # se quitan opciones y configuraciones de imagenes.
        try:
            pos = img_url.index("?")
            img_url = img_url[:pos]
        except ValueError:
            pass
        # finalmente la url es valida
        if url_valida(img_url):
            urls.append(img_url)
    return urls
````

### **Descarga de las imagenes**

Ahora que tenemos una función que captura todas las URL de las imágenes, necesitamos una función para descargar archivos de la web con Python, la función que realiza estas tareas es la siguiente:


```python
def DescargarArchivos(url, pathname):
    """
    Descarga un archivo de acuerdo a una url descrita y lo pone en la carpeta `pathname`
    """
    # si la carpeta no existe, se convierte la ruta usando isdir
    if not os.path.isdir(pathname):
        os.makedirs(pathname)
    # descargar el objeto de la respuesta por fragmento, no inmediatamente
    response = requests.get(url, stream=True)
    # obtener el tamaño total del archivo
    file_size = int(response.headers.get("Content-Length", 0))
    # obtener el nombre del archivo
    filename = os.path.join(pathname, url.split("/")[-1])
    # 
    # barra de progreso, cambiando la unidad a bytes en lugar de iteración (predeterminado por tqdm)
    progress = tqdm(response.iter_content(1024), f"Descargando {filename}", total=file_size, unit="B", unit_scale=True, unit_divisor=1024)
    with open(filename, "wb") as f:
        for data in progress.iterable:
            # escribir datos leídos en el archivo
            f.write(data)
            # actualizar la barra de progreso manualmente
            progress.update(len(data))
```
La función anterior básicamente toma la URL del archivo para descargar y el nombre de la ruta de la carpeta para guardar ese archivo.

````python
def DescargarArchivos(url, pathname):
    """
    Descarga un archivo de acuerdo a una url descrita y lo pone en la carpeta `pathname`
    """
    # si la carpeta no existe, se convierte la ruta usando isdir
    if not os.path.isdir(pathname):
        os.makedirs(pathname)
    # descargar el objeto de la respuesta por fragmento, no inmediatamente
    response = requests.get(url, stream=True)
    # obtener el tamaño total del archivo
    file_size = int(response.headers.get("Content-Length", 0))
    # obtener el nombre del archivo
    filename = os.path.join(pathname, url.split("/")[-1])
    # 
    # barra de progreso, cambiando la unidad a bytes en lugar de iteración (predeterminado por tqdm)
    progress = tqdm(response.iter_content(1024), f"Descargando {filename}", total=file_size, unit="B", unit_scale=True, unit_divisor=1024)
    with open(filename, "wb") as f:
        for data in progress.iterable:
            # escribir datos leídos en el archivo
            f.write(data)
            # actualizar la barra de progreso manualmente
            progress.update(len(data))
````

````python
def main(url, path):
    # Obtener todas las imagenes
    imgs = Todas_Las_Imagenes(url)
    i = 0
    for img in imgs:
        # para cada imagen descarguela.
        DescargarArchivos(img, path)
        i += 1
    print('Proceso Terminado, un total de {} imagenes procesadas'.format(i))
````

## **Ejercicio 10**

Escriba un programa Python para descargar los 250 datos principales de las mejores peliculas de IMDB (nombre de la película, lanzamiento inicial, nombre del director y estrellas) de acuerdo al vínculo [TopMovies](http://www.imdb.com/chart/top)

````python
from bs4 import BeautifulSoup
import requests
import re

# Descargar los datos
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

#Establecer la conexion con los dtos y los encabezados con los tags HTML
movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
votes = [b.attrs.get('data-value') for b in soup.select('td.ratingColumn strong')]
#Creamos una lista vacía
imdb = []

# Guardamos cada elemento en el diccionario (datos), luego en una lista (imdb)
for index in range(0, len(movies)):
    # Separamos los datos
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "star_cast": crew[index],
            "rating": ratings[index],
            "vote": votes[index],
            "link": links[index]}
    imdb.append(data)

for item in imdb:
    print(item['place'], '-', item['movie_title'], '('+item['year']+') -', 'Starring:', item['star_cast'])
  
````
